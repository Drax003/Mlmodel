## Conduct Exploratory Data Analysis (EDA)

### Subtask:
Perform a comprehensive exploratory data analysis to understand the distribution, relationships, and characteristics of the features in the housing dataset. This includes generating descriptive statistics and creating various visualizations to gain insights into the data.

````
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the California housing dataset
housing = fetch_california_housing(as_frame=False)

# Create a pandas DataFrame from the dataset's features
df = pd.DataFrame(housing.data, columns=housing.feature_names)

# Add the target variable (housing prices) as a new column
df['MedHouseVal'] = housing.target

# Display the first few rows of the DataFrame
print("California Housing DataFrame loaded successfully.")
print("Shape of the DataFrame:", df.shape)
df.head()
````

## Perform Data Cleaning and Preprocessing

### Subtask:
Clean the loaded housing data by handling missing values (e.g., imputation or removal), encoding categorical features if any, and splitting the data into training and testing sets. Feature scaling might also be applied if necessary.


**Reasoning**:
The user has provided the instructions for the EDA subtask in a markdown block. I will now generate a code block to execute these instructions.



````
import matplotlib.pyplot as plt
import seaborn as sns
import math

# 1. Generate descriptive statistics
print("\nDescriptive Statistics for the Housing Dataset:")
print(df.describe())

# 2. Create histograms for each numerical feature
print("\nGenerating Histograms for each feature...")
num_features = df.shape[1]
num_cols = 3 # Number of columns for subplots
num_rows = math.ceil(num_features / num_cols)

plt.figure(figsize=(num_cols * 5, num_rows * 4))
for i, column in enumerate(df.columns):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# 3. Generate scatter plots of each feature against the target variable
print("\nGenerating Scatter Plots of features vs. MedHouseVal...")
features = df.columns.drop('MedHouseVal')
num_features = len(features)
num_cols = 3 # Number of columns for subplots
num_rows = math.ceil(num_features / num_cols)

plt.figure(figsize=(num_cols * 5, num_rows * 4))
for i, column in enumerate(features):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.scatterplot(x=df[column], y=df['MedHouseVal'])
    plt.title(f'{column} vs. MedHouseVal')
plt.tight_layout()
plt.show()

# 4. Create a correlation matrix heatmap
print("\nGenerating Correlation Matrix Heatmap...")
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features and Target')
plt.show()

print("Exploratory Data Analysis complete.")
````
